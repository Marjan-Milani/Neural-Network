```python
import numpy as np
import matplotlib.pyplot as plt
```

<div style="font-family:Tahoma; font-size:16px; direction:rtl; color:Blue;">
داخل متغیر X عددی تصادفی را ضرب در اعداد موجود کن 
<br>
داخل متغیر my noise اندازه ی متغیر های X  را ضرب در اعدادی تصادفی کن و بریز
<br>
سینوس متغیر X را در my noise جمع کرده و داخل Y بریز
    </div> 

```python
X = 6*np.random.rand(1,1000)-3
my_noise=0.2*np.random.randn(np.size(X))
y = np.sin(X)+my_noise
```

<div style="font-family:Tahoma; font-size:16px; direction:rtl; color:Blue;">
نمودار را بساز و نمایش بده
    </div> 

```python
plt.plot(X,y,'.')

plt.show()
```
